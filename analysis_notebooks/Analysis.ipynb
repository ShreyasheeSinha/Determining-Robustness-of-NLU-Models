```python
def read_jsonl_file(path):
  with open(path, 'r') as json_file:
      json_list = list(json_file)

  results = []
  for json_str in json_list:
      results.append(json.loads(json_str))
  return results
```

```python
test_data = pd.DataFrame(read_jsonl_file("/content/gdrive/MyDrive/paraphraser_work/Final_datasets/rte_test_p.jsonl"))
test_data = test_data.fillna(value=np.nan)
```

```python
predictions = pd.read_csv('bilstm_3_class_rte_test_p_predictions.csv')
predictions
```

```python
# THIS IS KEY TO OUR WORK, DO NOT ALTER
df = pd.DataFrame()

for index, row in test_data.iterrows():
  dataset = row['dataset']
  corpus_sent_id = row['corpus_sent_id']
  s1_para_id = row['s1_para_id']
  s2_para_id = row['s2_para_id']
  prediction_rows = predictions[(predictions['dataset'] == dataset) & (predictions['corpus_sent_id'] == corpus_sent_id)]
  if pd.isnull(s1_para_id) and pd.isnull(s2_para_id):
    prediction_rows = prediction_rows[(prediction_rows['s1_para_id'].isnull()) & (prediction_rows['s2_para_id'].isnull())]
  elif pd.isnull(s1_para_id):
    prediction_rows = prediction_rows[(prediction_rows['s1_para_id'].isnull()) & (prediction_rows['s2_para_id'] == s2_para_id)]
  elif pd.isnull(s2_para_id):
    prediction_rows = prediction_rows[(prediction_rows['s1_para_id'] == s1_para_id) & (prediction_rows['s2_para_id'].isnull())]
  else:
    prediction_rows = prediction_rows[(prediction_rows['s1_para_id'] == s1_para_id) & (prediction_rows['s2_para_id'] == s2_para_id)]
  df = df.append(prediction_rows)
```

```python
ph = df.copy()
ph_ = df.copy()
p_h = df.copy()
p_h_ = df.copy()

# ph = ph.loc[ph['s1_para_id'].isnull() & ph['s2_para_id'].isnull()]
# p_h_ = p_h_.loc[~(p_h_['s1_para_id'].isnull() & p_h_['s2_para_id'].isnull())]
# # p_h = p_h.loc[~p_h['s1_para_id'].isnull() & p_h['s2_para_id'].isnull()]
# # p_h_ = p_h_.loc[~p_h_['s1_para_id'].isnull() & ~p_h_['s2_para_id'].isnull()]
ph = ph.loc[ph['s1_para_id'].isnull() & ph['s2_para_id'].isnull()]
p_h = p_h.loc[~p_h['s1_para_id'].isnull() & p_h['s2_para_id'].isnull()]
ph_ = ph_.loc[ph_['s1_para_id'].isnull() & ~ph_['s2_para_id'].isnull()]
p_h_ = p_h_.loc[~p_h_['s1_para_id'].isnull() & ~p_h_['s2_para_id'].isnull()]
```

```python
either_paraphrased = pd.concat([p_h, ph_])
either_paraphrased
```

```python
def get_task_wise_split(df):
  IR = df.copy()
  QA = df.copy()
  SUM = df.copy()
  IE = df.copy()
  RC = df.copy()
  CD = df.copy()
  MT = df.copy()
  PP = df.copy()

  IR = IR[IR['task'] == 'IR']
  QA = QA[QA['task'] == 'QA']
  SUM = SUM[SUM['task'] == 'SUM']
  IE = IE[IE['task'] == 'IE']
  RC = RC[RC['task'] == 'RC']
  CD = CD[CD['task'] == 'CD']
  MT = MT[MT['task'] == 'MT']
  PP = PP[PP['task'] == 'PP']

  return [IR, QA, SUM, IE, RC, CD, MT, PP]
```

```python
ph_stats = pd.DataFrame(columns = ['Accuracy', 'Precision', 'Recall', 'F1'])
either_paraphrased_stats = pd.DataFrame(columns = ['Accuracy', 'Precision', 'Recall', 'F1'])
p_h_stats = pd.DataFrame(columns = ['Accuracy', 'Precision', 'Recall', 'F1'])
```

```python
# from sklearn.metrics import precision_score, recall_score, f1_score
# def stats(df):
#   acc = np.sum(df['prediction'] == df['gold_label']) / len(df['gold_label'])
#   precision = precision_score(df['gold_label'], df['prediction'], zero_division=0)
#   recall = recall_score(df['gold_label'], df['prediction'], zero_division=0)
#   f1 = f1_score(df['gold_label'], df['prediction'], zero_division=0)

#   print("Accuracy : ", acc)
#   print("Precision : ", precision)
#   print("Recall : ", recall)
#   print("F1 : ", f1)
#   print(" ")

from sklearn.metrics import precision_score, recall_score, f1_score
def stats(df, res):
  acc = np.sum(df['prediction'] == df['gold_label']) / len(df['gold_label'])
  precision = precision_score(df['gold_label'], df['prediction'], zero_division=0)
  recall = recall_score(df['gold_label'], df['prediction'], zero_division=0)
  f1 = f1_score(df['gold_label'], df['prediction'], zero_division=0)
  temp_dict = {
      "Accuracy": acc * 100,
      "Precision": precision,
      "Recall": recall,
      "F1": f1
  }
  res = res.append(temp_dict, ignore_index = True)

  print("Accuracy : ", acc)
  print("Precision : ", precision)
  print("Recall : ", recall)
  print("F1 : ", f1)
  print(" ")
  return res
```
