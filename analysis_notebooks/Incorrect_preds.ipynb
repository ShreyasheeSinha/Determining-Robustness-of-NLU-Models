```python
import pandas as pd
import json
import numpy as np
SHEET_NAME = "roberta-mlp"
```

```python
def read_jsonl_file(path):
  with open(path, 'r') as json_file:
      json_list = list(json_file)

  results = []
  for json_str in json_list:
      results.append(json.loads(json_str))
  return results
```

```python
test_data = pd.DataFrame(read_jsonl_file("/content/gdrive/MyDrive/paraphraser_work/Final_datasets/rte_test_p.jsonl")) # Unannotated dataset: /content/gdrive/MyDrive/paraphraser_work/MTurk data/Generated_dataset/rte_test_paraphrased_p_d_or_h_d_sorted.jsonl
test_data = test_data.fillna(value=np.nan)
```

```python
predictions = pd.read_csv('roberta_large_3_class_rte_test_p_predictions.csv')
predictions
```

```python
# THIS IS KEY TO OUR WORK, DO NOT ALTER
df = pd.DataFrame()

for index, row in test_data.iterrows():
  dataset = row['dataset']
  corpus_sent_id = row['corpus_sent_id']
  s1_para_id = row['s1_para_id']
  s2_para_id = row['s2_para_id']
  prediction_rows = predictions[(predictions['dataset'] == dataset) & (predictions['corpus_sent_id'] == corpus_sent_id)]
  if pd.isnull(s1_para_id) and pd.isnull(s2_para_id):
    prediction_rows = prediction_rows[(prediction_rows['s1_para_id'].isnull()) & (prediction_rows['s2_para_id'].isnull())]
  elif pd.isnull(s1_para_id):
    prediction_rows = prediction_rows[(prediction_rows['s1_para_id'].isnull()) & (prediction_rows['s2_para_id'] == s2_para_id)]
  elif pd.isnull(s2_para_id):
    prediction_rows = prediction_rows[(prediction_rows['s1_para_id'] == s1_para_id) & (prediction_rows['s2_para_id'].isnull())]
  else:
    prediction_rows = prediction_rows[(prediction_rows['s1_para_id'] == s1_para_id) & (prediction_rows['s2_para_id'] == s2_para_id)]
  df = df.append(prediction_rows)
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
df
```

```python
def task_r1(df):
  grouped_df = df.groupby(["dataset", "corpus_sent_id"])
  incorrect_preds_df = pd.DataFrame()
  for name, group in grouped_df:
    ph = group[group['s1_para_id'].isnull() & group['s2_para_id'].isnull()]
    ph_pred = group[group['s1_para_id'].isnull() & group['s2_para_id'].isnull()]["prediction"].to_list()[0]
    paraphrased_preds = group[~(group['s1_para_id'].isnull() & group['s2_para_id'].isnull())]
    for index, row in paraphrased_preds.iterrows():
      if row["prediction"] != ph_pred:
        # incorrect_preds_df = incorrect_preds_df.append(row)
        dataframe_row = {}
        original_dataset_corpus_id = ph.iloc[0]["dataset"] + "_" + ph.iloc[0]["corpus_sent_id"].astype(str)
        original_premise = ph.iloc[0]["sentence1"]
        original_hypothesis = ph.iloc[0]["sentence2"]
        original_gold_label = ph.iloc[0]["gold_label"]
        original_prediction = ph.iloc[0]["prediction"]
        paraphrased_premise_id = row["s1_para_id"]
        paraphrased_premise = row["sentence1"]
        paraphrased_hypothesis_id = row["s2_para_id"]
        paraphrased_hypothesis = row["sentence2"]
        paraphrased_gold_label = row["gold_label"]
        paraphrase_prediction = row["prediction"]
        original_pair = "P: " + original_premise + "\nH: " + original_hypothesis
        paraphrased_pair = "P: " + paraphrased_premise + "\nH: " + paraphrased_hypothesis
        dataframe_row["dataset_corpus_id"] = original_dataset_corpus_id
        dataframe_row["original_pair"] = original_pair
        dataframe_row["gold_label"] = original_gold_label
        dataframe_row["paraphrased_pair"] = paraphrased_pair
        dataframe_row["s1_para_id"] = paraphrased_premise_id
        dataframe_row["s2_para_id"] = paraphrased_hypothesis_id
        dataframe_row["original_pair_prediction"] = original_prediction
        dataframe_row["paraphrased_pair_prediction"] = paraphrase_prediction
        incorrect_preds_df = incorrect_preds_df.append(dataframe_row, ignore_index=True)
  return incorrect_preds_df
```

```python
incorrect_preds = task_r1(df)
```
