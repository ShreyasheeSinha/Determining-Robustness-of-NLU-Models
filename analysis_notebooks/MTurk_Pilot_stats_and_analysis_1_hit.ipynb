## Reference for HTML and answer columns

1.   Answer.gm{j}: The reference sentence is not grammatical. 1 means it is not grammatical (j = 0, 2, 4, 6).
2.   Answer.gm{i}: The selected answer for the ith given paraphrase (i = 1, 3, 5, 7). The following are the values it can take and their meaning. It can take 3 values:

> *   2 - Indicates that the given paraphrase is fluent and grammatical,
*   1 - Indicates that the given paraphrase contains grammatical errors (besides casing & extra spaces),
*   0 - Indicates that the given paraphrase makes absolutely no sense and the user cannot comment on its meaning.

3. Answer.pg{j}:  The reference sentence contains offensive contents. 1 means it contains offensive contents (j = 0, 2, 4, 6).

4. Answer.range{i}: The selected slider answer value for the ith given paraphrase (i = 1, 3, 5, 7). A score closer to 100 implies that a given paraphrase is more identical in meaning to the reference sentence. A score closer to 0 implies that a given paraphrase is less identical in meaning to the reference sentence.

```python
columns = ['Input.sen1_origin_id', 'Input.sentence1', 'Input.sen1_paraphrase0', 'Input.sen2_origin_id', 'Input.sentence2', 'Input.sen2_paraphrase0', 'Input.sen3_origin_id', 'Input.sentence3', 'Input.sen3_paraphrase0', 'Input.sen4_origin_id', 'Input.sentence4', 'Input.sen4_paraphrase0', 'Answer.gm0', 'Answer.gm1', 'Answer.gm2', 'Answer.gm3', 'Answer.gm4', 'Answer.gm5', 'Answer.gm6', 'Answer.gm7', 'Answer.pg0', 'Answer.pg2', 'Answer.pg4', 'Answer.pg6', 'Answer.range1', 'Answer.range3', 'Answer.range5', 'Answer.range7']
filtered_data = pilot_data[columns]
filtered_data.head()
```

```python
for i in range(0, 7, 2):
  column_name = 'Answer.gm' + str(i)
  number_of_grammatically_incorrect_reference_sentences = len(filtered_data[filtered_data[column_name] == '1'])
  print("Number of grammatically incorrect reference sentences:", number_of_grammatically_incorrect_reference_sentences)
  column_name = 'Answer.pg' + str(i)
  number_of_offensive_reference_sentences = len(filtered_data[filtered_data[column_name] == '1'])
  print("Number of offensive reference sentences:", number_of_offensive_reference_sentences)
```

```python
sen1_paraphrase1_data = filtered_data[['Input.sen1_origin_id', 'Input.sentence1', 'Input.sen1_paraphrase0', 'Answer.gm1', 'Answer.range1']]
sen2_paraphrase1_data = filtered_data[['Input.sen2_origin_id', 'Input.sentence2', 'Input.sen2_paraphrase0', 'Answer.gm3', 'Answer.range3']]
sen3_paraphrase1_data = filtered_data[['Input.sen3_origin_id', 'Input.sentence3', 'Input.sen3_paraphrase0', 'Answer.gm5', 'Answer.range5']]
sen4_paraphrase1_data = filtered_data[['Input.sen4_origin_id', 'Input.sentence4', 'Input.sen4_paraphrase0', 'Answer.gm7', 'Answer.range7']]
```

```python
def get_number_of_type_sentences(data, paraphrase_number, answer_type):
  column_name = 'Answer.gm' + str(paraphrase_number)
  error_data = data[data[column_name] == answer_type]
  return error_data
```

```python
def get_sentences_above_threshold(data, paraphrase_number, threshold):
  column_name = 'Answer.range' + str(paraphrase_number)
  return data[data[column_name] >= threshold]
```

```python
THRESHOLD = 51
```

```python
s1_p1_grammar_correct_sentences = get_number_of_type_sentences(sen1_paraphrase1_data, 1, 2)
s1_p1_grammar_correct_thresholded_sentences = get_sentences_above_threshold(s1_p1_grammar_correct_sentences, 1, THRESHOLD)
```

```python
s2_p1_grammar_correct_sentences = get_number_of_type_sentences(sen2_paraphrase1_data, 3, 2)
s2_p1_grammar_correct_thresholded_sentences = get_sentences_above_threshold(s2_p1_grammar_correct_sentences, 3, THRESHOLD)
```

```python
s3_p1_grammar_correct_sentences = get_number_of_type_sentences(sen3_paraphrase1_data, 5, 2)
s3_p1_grammar_correct_thresholded_sentences = get_sentences_above_threshold(s3_p1_grammar_correct_sentences, 5, THRESHOLD)
```

```python
s4_p1_grammar_correct_sentences = get_number_of_type_sentences(sen4_paraphrase1_data, 7, 2)
s4_p1_grammar_correct_thresholded_sentences = get_sentences_above_threshold(s4_p1_grammar_correct_sentences, 7, THRESHOLD)
```

```python
all_grammatically_correct_paraphrases = pd.DataFrame( np.concatenate((s1_p1_grammar_correct_sentences.values, s2_p1_grammar_correct_sentences.values, s3_p1_grammar_correct_sentences.values, s4_p1_grammar_correct_sentences.values), axis=0 ))
all_grammatically_correct_paraphrases.columns = ['origin_id', 'sentence', 'paraphrase0', 'gm', 'range']
all_grammatically_correct_paraphrases
```

```python
all_grammatically_correct_thresholded_paraphrases = pd.DataFrame( np.concatenate((s1_p1_grammar_correct_thresholded_sentences.values, s2_p1_grammar_correct_thresholded_sentences.values, s3_p1_grammar_correct_thresholded_sentences.values, s4_p1_grammar_correct_thresholded_sentences.values), axis=0 ))
all_grammatically_correct_thresholded_paraphrases.columns = ['origin_id', 'sentence', 'paraphrase0', 'gm', 'range']
all_grammatically_correct_thresholded_paraphrases
```
