## Reference for HTML and answer columns

1.   Answer.gm0: The reference sentence is not grammatical. 1 means it is not grammatical.
2.   Answer.gm{i}: The selected answer for the ith given paraphrase (i = 1, 2, 3). The following are the values it can take and their meaning. It can take 3 values:

> *   2 - Indicates that the given paraphrase is fluent and grammatical,
*   1 - Indicates that the given paraphrase contains grammatical errors (besides casing & extra spaces),
*   0 - Indicates that the given paraphrase makes absolutely no sense and the user cannot comment on its meaning.

3. Answer.pg0:  The reference sentence contains offensive contents. 1 means it contains offensive contents.

4. Answer.range{i}: The selected slider answer value for the ith given paraphrase (i = 1, 2, 3). A score closer to 100 implies that a given paraphrase is more identical in meaning to the reference sentence. A score closer to 0 implies that a given paraphrase is less identical in meaning to the reference sentence.

5. Answer.gm4: The reference sentence is not grammatical. 1 means it is not grammatical.

6. Answer.gm5: The selected answer for the given paraphrase. The following are the values it can take and their meaning. It can take 3 values:

> *   2 - Indicates that the given paraphrase is fluent and grammatical,
*   1 - Indicates that the given paraphrase contains grammatical errors (besides casing & extra spaces),
*   0 - Indicates that the given paraphrase makes absolutely no sense and the user cannot comment on its meaning.

7. Answer.pg4:  The reference sentence contains offensive contents. 1 means it contains offensive contents.

8. Answer.range5: The selected slider answer value for the given paraphrase. A score closer to 100 implies that a given paraphrase is more identical in meaning to the reference sentence. A score closer to 0 implies that a given paraphrase is less identical in meaning to the reference sentence.

```python
sen1_paraphrase1_data = filtered_data[['Input.sen1_origin_id', 'Input.sentence1', 'Input.sen1_paraphrase0', 'Answer.gm1', 'Answer.range1']]
sen1_paraphrase2_data = filtered_data[['Input.sen1_origin_id', 'Input.sentence1', 'Input.sen1_paraphrase1', 'Answer.gm2', 'Answer.range2']]
sen1_paraphrase3_data = filtered_data[['Input.sen1_origin_id', 'Input.sentence1', 'Input.sen1_paraphrase2', 'Answer.gm3', 'Answer.range3']]
sen2_paraphrase1_data = filtered_data[['Input.sen2_origin_id', 'Input.sentence2', 'Input.sen2_paraphrase0', 'Answer.gm5', 'Answer.range5']]
```

```python
def get_number_of_type_sentences(data, paraphrase_number, answer_type):
  column_name = 'Answer.gm' + str(paraphrase_number)
  error_data = data[data[column_name] == answer_type]
  return error_data
```

```python
def get_sentences_above_threshold(data, paraphrase_number, threshold):
  column_name = 'Answer.range' + str(paraphrase_number)
  return data[data[column_name] >= threshold]
```

```python
THRESHOLD = 51
```

```python
s1_p1_grammar_correct_sentences = get_number_of_type_sentences(sen1_paraphrase1_data, 1, 2)
s1_p1_grammar_correct_thresholded_sentences = get_sentences_above_threshold(s1_p1_grammar_correct_sentences, 1, THRESHOLD)
s1_p2_grammar_correct_sentences = get_number_of_type_sentences(sen1_paraphrase2_data, 2, 2)
s1_p2_grammar_correct_thresholded_sentences = get_sentences_above_threshold(s1_p2_grammar_correct_sentences, 2, THRESHOLD)
s1_p3_grammar_correct_sentences = get_number_of_type_sentences(sen1_paraphrase3_data, 3, 2)
s1_p3_grammar_correct_thresholded_sentences = get_sentences_above_threshold(s1_p3_grammar_correct_sentences, 3, THRESHOLD)
```

```python
s2_p1_grammar_correct_sentences = get_number_of_type_sentences(sen2_paraphrase1_data, 5, 2)
s2_p1_grammar_correct_thresholded_sentences = get_sentences_above_threshold(s2_p1_grammar_correct_sentences, 5, THRESHOLD)
```

```python
def get_para_counts(data):
  ONE_PARAPHRASES = "one_para"
  TWO_PARAPHRASES = "two_para"
  THREE_PARAPHRASES = "three_para"
  para_count_dict = {
      ONE_PARAPHRASES: 0,
      TWO_PARAPHRASES: 0,
      THREE_PARAPHRASES: 0
  }
  for index, row in data.iterrows():
    para1 = row['Input.paraphrase0']
    para2 = row['Input.paraphrase1']
    para3 = row['Input.paraphrase2']
    if (len(para1) > 0 and len(para2) > 0 and len(para3) > 0):
      para_count_dict[THREE_PARAPHRASES] += 1
    elif (len(para1) > 0 and len(para2) > 0) or (len(para2) > 0 and len(para3) > 0) or (len(para3) > 0 and len(para1) > 0):
      para_count_dict[TWO_PARAPHRASES] += 1
    else:
      para_count_dict[ONE_PARAPHRASES] += 1
  return para_count_dict
```
