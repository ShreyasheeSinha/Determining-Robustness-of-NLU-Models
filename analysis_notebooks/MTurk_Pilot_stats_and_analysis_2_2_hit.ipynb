## Reference for HTML and answer columns

1.   Answer.gm0: The reference sentence is not grammatical. 1 means it is not grammatical.
2.   Answer.gm{i}: The selected answer for the ith given paraphrase (i = 1, 2). The following are the values it can take and their meaning. It can take 3 values:

> *   2 - Indicates that the given paraphrase is fluent and grammatical,
*   1 - Indicates that the given paraphrase contains grammatical errors (besides casing & extra spaces),
*   0 - Indicates that the given paraphrase makes absolutely no sense and the user cannot comment on its meaning.

3. Answer.pg0:  The reference sentence contains offensive contents. 1 means it contains offensive contents.

4. Answer.range{i}: The selected slider answer value for the ith given paraphrase (i = 1, 2). A score closer to 100 implies that a given paraphrase is more identical in meaning to the reference sentence. A score closer to 0 implies that a given paraphrase is less identical in meaning to the reference sentence.

5. Answer.gm3: The reference sentence is not grammatical. 1 means it is not grammatical.

6. Answer.gm{j}: The selected answer for the given paraphrase (j = 4, 5). The following are the values it can take and their meaning. It can take 3 values:

> *   2 - Indicates that the given paraphrase is fluent and grammatical,
*   1 - Indicates that the given paraphrase contains grammatical errors (besides casing & extra spaces),
*   0 - Indicates that the given paraphrase makes absolutely no sense and the user cannot comment on its meaning.

7. Answer.pg3:  The reference sentence contains offensive contents. 1 means it contains offensive contents.

8. Answer.range{j}: The selected slider answer value for the given paraphrase (j = 4, 5). A score closer to 100 implies that a given paraphrase is more identical in meaning to the reference sentence. A score closer to 0 implies that a given paraphrase is less identical in meaning to the reference sentence.

```python
columns = ['Input.sen1_origin_id', 'Input.sentence1', 'Input.sen1_paraphrase0', 'Input.sen1_paraphrase1', 'Input.sen2_origin_id', 'Input.sentence2', 'Input.sen2_paraphrase0', 'Input.sen2_paraphrase1', 'Answer.gm0', 'Answer.gm1', 'Answer.gm2', 'Answer.gm3', 'Answer.gm4', 'Answer.gm5', 'Answer.pg0', 'Answer.pg3', 'Answer.range1', 'Answer.range2', 'Answer.range4', 'Answer.range5']
filtered_data = pilot_data[columns]
filtered_data.head()
```

```python
sen1_paraphrase1_data = filtered_data[['Input.sen1_origin_id', 'Input.sentence1', 'Input.sen1_paraphrase0', 'Answer.gm1', 'Answer.range1']]
sen1_paraphrase2_data = filtered_data[['Input.sen1_origin_id', 'Input.sentence1', 'Input.sen1_paraphrase1', 'Answer.gm2', 'Answer.range2']]
sen2_paraphrase1_data = filtered_data[['Input.sen2_origin_id', 'Input.sentence2', 'Input.sen2_paraphrase0', 'Answer.gm4', 'Answer.range4']]
sen2_paraphrase2_data = filtered_data[['Input.sen2_origin_id', 'Input.sentence2', 'Input.sen2_paraphrase1', 'Answer.gm5', 'Answer.range5']]
```

```python
def get_number_of_type_sentences(data, paraphrase_number, answer_type):
  column_name = 'Answer.gm' + str(paraphrase_number)
  error_data = data[data[column_name] == answer_type]
  return error_data
```

```python
def get_sentences_above_threshold(data, paraphrase_number, threshold):
  column_name = 'Answer.range' + str(paraphrase_number)
  return data[data[column_name] >= threshold]
```

```python
THRESHOLD = 51
```

```python
s1_p1_grammar_correct_sentences = get_number_of_type_sentences(sen1_paraphrase1_data, 1, 2)
s1_p1_grammar_correct_thresholded_sentences = get_sentences_above_threshold(s1_p1_grammar_correct_sentences, 1, THRESHOLD)
s1_p2_grammar_correct_sentences = get_number_of_type_sentences(sen1_paraphrase2_data, 2, 2)
s1_p2_grammar_correct_thresholded_sentences = get_sentences_above_threshold(s1_p2_grammar_correct_sentences, 2, THRESHOLD)

s2_p1_grammar_correct_sentences = get_number_of_type_sentences(sen2_paraphrase1_data, 4, 2)
s2_p1_grammar_correct_thresholded_sentences = get_sentences_above_threshold(s2_p1_grammar_correct_sentences, 4, THRESHOLD)
s2_p2_grammar_correct_sentences = get_number_of_type_sentences(sen2_paraphrase2_data, 5, 2)
s2_p2_grammar_correct_thresholded_sentences = get_sentences_above_threshold(s2_p2_grammar_correct_sentences, 5, THRESHOLD)
```

```python
s1_all_grammatically_correct_paraphrases = pd.merge(s1_p1_grammar_correct_sentences, s1_p2_grammar_correct_sentences, how="outer", on=["Input.sen1_origin_id", "Input.sentence1"])
s1_all_grammatically_correct_paraphrases
```

```python
s1_all_grammatically_correct_thresholded_paraphrases = pd.merge(s1_p1_grammar_correct_thresholded_sentences, s1_p2_grammar_correct_thresholded_sentences, how="outer", on=["Input.sen1_origin_id", "Input.sentence1"])
s1_all_grammatically_correct_thresholded_paraphrases
```

```python
s2_all_grammatically_correct_paraphrases = pd.merge(s2_p1_grammar_correct_sentences, s2_p2_grammar_correct_sentences, how="outer", on=["Input.sen2_origin_id", "Input.sentence2"])
s2_all_grammatically_correct_paraphrases
```

```python
s2_all_grammatically_correct_thresholded_paraphrases = pd.merge(s2_p1_grammar_correct_thresholded_sentences, s2_p2_grammar_correct_thresholded_sentences, how="outer", on=["Input.sen2_origin_id", "Input.sentence2"])
s2_all_grammatically_correct_thresholded_paraphrases
```

```python
all_grammatically_correct_paraphrases = pd.DataFrame( np.concatenate((s1_all_grammatically_correct_paraphrases.values, s2_all_grammatically_correct_paraphrases.values), axis=0 ))
all_grammatically_correct_paraphrases
```

```python
all_grammatically_correct_paraphrases.columns = ['origin_id', 'sentence', 'paraphrase0', 'gm1', 'range1', 'paraphrase1', 'gm1', 'range1']
all_grammatically_correct_paraphrases
```
